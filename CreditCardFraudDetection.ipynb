
# AI program that detects credit fraud 

This dataset contains anonymized features of transactions, where some transactions are labeled as fraudulent or legitimate.

I have machine learning program using Python and popular libraries like pandas, sklearn, and xgboost for fraud detection.

The kaggle dataset is being used.

Here are the 4 steps in the program

 1. Load and preprocess the dataset.

2. Train a machine learning model.

3. Evaluate the model.

4. Deploy or use the model for prediction.


```python
#Load and Preprocess the Data
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Load the dataset
```

```python
data=pd.read_csv(r"D:\AI\creditcard.csv")
data
```

```python
# Display the first few rows of the dataset
```

```python
print(data.head())
```

```python
#Display the column info
```

```python
data.info()
```

```python
# Split the data into features and target
```

```python
X = data.drop(columns=['Class'])  # Features (everything except the target column)
y = data['Class']  # Target (the 'Class' column which indicates fraud)
```

```python
# Normalize the data 
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the dataset into training and testing sets
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

```python
#Train the Model
# Initialize and train a Random Forest classifier 
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
```

```python
model.fit(X_train, y_train)
```

```python
#Evaluate the Model
# Predict on the test set
```

```python
y_pred = model.predict(X_test)
```

```python
# Print the confusion matrix and classification report
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
 #Evaluate with Other Algorithms
#import the libraries
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```python
# Train an XGBoost model
```

```python
xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)
```

```python
# Predict on the test set
```

```python
y_pred_xgb = xgb_model.predict(X_test)
```
